```
src = requests.get("https://www.amazon.in/Sapiens-Humankind-Yuval-Noah-Harari/dp/0099590085?pf_rd_p=f8c448da-0b07-5b46-9376-30ec20c8f937&pf_rd_r=13BCY2AX9V4VQE4QN8T4&pd_rd_wg=h1QzI&ref_=pd_gw_ri&pd_rd_w=jvyqB&pd_rd_r=a61bafd7-56be-4b2f-b1a8-fecc01eee0b3", headers ={'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'}).text
soup = BeautifulSoup(src, 'lxml')
print(soup.prettify())

```

```
reviews = soup.find('div',attrs =  {"id" : 'cm-cr-dp-review-list', "class" : 'a-section review-views celwidget'})

#print(reviews)

reviews_text = reviews.find_all('div', attrs = {"class" : "a-row a-spacing-small review-data"})
reviews_collector = []
for review in reviews_text:
  a = review.text
  #print(a)
  a = a[:-10]
  #print(a)
  reviews_collector.append(a)

for review in reviews_collector:
  print(review)
  print()
#print(reviews_collector)
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

final_reviews = []
tagged_reviews = []
lemmatized_sentences = []
print()
print("---------- Customer Reviews from Amazon preprocessed ----------")
print()
for review in reviews_collector:
  word_tokens = word_tokenize(review)
  final_sentence = [w for w in word_tokens if not w in stop_words]
  final_reviews.append(final_sentence)
  lemmatized_sentence = [lemmatizer.lemmatize(w) for w in final_sentence]
  lemmatized_sentences.append(lemmatized_sentence)
  #lemmatizer.lemmatize()
  tagged = nltk.pos_tag(lemmatized_sentence)
  print(tagged)
  print()
  tagged_reviews.append(tagged)
```
